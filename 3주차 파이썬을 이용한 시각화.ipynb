```python
#  데이터 프레임 사용
import pandas as pd

df = pd.DataFrame([100, 150, 200, 250, 300], columns=['numbers'], index=['a', 'b', 'c', 'd', 'e'])


print(df.sum)# DataFrame의 Value를 모두 더함

print()

print(df**2) # numbers 칼럼의 각 값을 제곱

print()

# DataFrame에 value라는 컬럼을 추가하고 값을 입력
 # 데이터프레임 컬럼 추가 시 데이터 개수는 기존의 다른 컬럼의 데이터 개수와 동일해야 함 (!!)
df['values'] = (10, 50, 40, 30, 60)
print(df)

print()

# DataFrame의 Column 추가 (오류발생!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!)

df['values']=pd.DataFrame(['Fourth', 'Second', 'First', 'Fifth', 'Third'], 
                         index=['d', 'b', 'a', 'e', 'c'])
print(df)
print()


# DataFrame의 Column 삭제
del df['values']
print(df)

# DataFrame의 데이터 개수(index) 일치(매우 중요!!!!!!!)
    #  Length of values does not match length of index
# df['values'] = (10, 50, 40)
# print(df)
print()
########################################
# DataFrame을 하나의 데이터프레임으로 합치는 방법
    # join, merge, concat

# Column A에 1,2,3 값 넣음
df_1 = pd.DataFrame(['1', '2', '3'], columns=['A'])

# Column B에 4,5,6,7 값 넣음
df_2 = pd.DataFrame(['4', '5', '6', '7'], columns = ['B'])

df = df_1.join(df_2, how='outer') # outer join을 의미
print(df)



```

```python
# 주안동  시민회관 4거리 창작홀 시간 11시 5시 
#########################################
# pandas 데이터프레임에서 임의의 숫자, 난수를 생성해서 최대, 최소갑승ㄹ 구하는 예제
    # numpy의 Random 함수
    
import pandas as pd
import numpy as np

# np.random.randn은 5행 5열의 랜덤 데이터를 생성
# 재호출 시 데이터가 변형됨
df = pd.DataFrame(np.random.randn(5,5))
df.columns=['A', 'B', 'C', 'D', 'E']
print(df)
print()
# DataFrame을 이용한 연산

    # 1. 최대값 max
print(df.max())
print()
    # 2. 최소값 min함수
print(df.min())
print()
    # 3. 평균값 mean 함수
print(df.mean())
print()
    # 4. 표준편차값
print(df.std())
print()
    # 5. 누적합 cumsum(칼럼의 누적값)
print(df.cumsum())
print()

# Describe 함수 Group by를 이용한 DataFrame
# descirbe 함수를 이용한 통계적 분포 확인
print(df.describe())


# Group by 하기전 그룹별로 구분하기 위해 division이라는 Column을 생성
df['division'] =  ['X', 'X', 'X', 'Y', 'Z']
print(df)

    # DataFrame의 Group by를 사용해 Column division의 값에 따라 평균값을 산출
df.groupby(['division']).mean()
```

```python
# 시각화 라이브러리 %matplotli
import matplotlib as mpl
import matplotlib.pyplot as plt 
import numpy as np # 행렬, 벡터 등의 수학 계산을 위한 자료구조와 계산 함수 제공 패키지

# 아이파이썬의 매직함수로서, matplotlib의 시각화 결과를 출력해줌.
%matplotlib inline 

# standard_normal() : 표준정규분포를 따르는 랜덤값을 생성하는 함수로서, 40개의 난수를 생성
# (nump.ndarray형 변수 즉, 다차원 배열형(!!)
value = np.random.standard_normal(40)
value

# 선 Plot 그리기 및 x/y축 범위 설정
index = range(len(value))

# 여기서 그림1이 출력됨
plt.plot(index, value) # 선이나 마커를 플롯하는 함수
    
plt.xlim(0, 20) # x축 범위 설정
plt.ylim(np.min(value) -1, np.min(value) + 1) # y축 범위 설정 value 변수의 최솟값보다 하나 작은 값으로 아래 한계를 정하고, 최대값보다 하나 큰 값으로 위의 한계를 정함

# Plot 옵션 설정 및 꾸미기
plt.figure(figsize=(7, 4)) # plot의 모양. 가로 세로가 7인치, 4인치 비로 설정
plt.plot(value.cumsum(), 'b', lw=1.5) # plot 함수 호출 numpy.cumsum() : plot의 새로운 figure을 생성하는 함수. 'b' : 파란색, w=1.5(표시할 스타일)
plt.plot(value.cumsum(), 'ro') # plot 함수 호출 'ro'는 빨간색 동그라미를 의미
plt.xlabel('index') # x축 이름 설정
plt.ylabel('value') # y축 이름 설정
plt.title("Line Plot1") # 플롯의 제목을 설정하는 함수

print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
```

```python
# matplolib & seborn & pandas


# 히스토그램의 빈(bin) 시각화 (데이터의 형상과 산포 정도를 평가하는 데 많이 사용)
    # pyplot.hist() 히스토그램을 그리는 함수
value2 = np.random.standard_normal((500, 2)) # 표준정규분표 형태의 난수 500개씩 2줄의 배열의 값
plt.figure(figsize=(12, 7))
plt.hist(value2, label=('1st', '2nd'), bins=30) # 두 개의 데이터 셋에 라벨을 붕팀, bins : 데이터를 구간으로 나누는 개수 (예러남 ㅆㅃㅆㅃㅆㅃㅆㅃ)
plt.grid(True) # 격자식
plt.legend(bbox_to_anchor=(1.2, 1), loc='upper right') # bbox_to_anchor : 라벨의 상대적인 위치를 잡아주기 위한 옵션, upper right : 맨오른쪽 상단에 위치
plt.xlabel('label')
plt.ylabel('frequency')
plt.title('Histogram 1')


# 히스토그램 색 입히기

plt.figure(figsize=(12, 6))
plt.hist(value2, label=('1st', '2nd'), bins=30,  color=['red', 'green']) # 두 개의 데이터 셋에 라벨을 붕팀, bins : 데이터를 구간으로 나누는 개수
plt.grid(True) # 격자식
plt.legend(bbox_to_anchor=(1.2, 1), loc='upper right') # bbox_to_anchor : 라벨의 상대적인 위치를 잡아주기 위한 옵션, upper right : 맨오른쪽 상단에 위치
plt.xlabel('label')
plt.ylabel('frequency')
plt.title('Histogram 1')

    #     각 patch에 다른 색 입히기 (patch는 직사각형 형태의 막대기를 의미함)
    #     for문 + setp() 함수를 이용한다.
    
value2=np.random.standard_normal(500)
cm = plt.cm.get_cmap('Spectral') # cm : colormap들중에서 인스턴스를 하나 가져옴
plt.figure(figsize=(12, 6))

    #     plot historgram (cm 컬러맵 변수느 0부터 1사이의 값을 바당서 이를 대응하는 색상으로 변환해주는 역할을 하기 때문에, 각 빈들의 중간값을 구한 다음에 값들을 0에서 1사이의 값으로  범위를 좁히는 작업이 필요)
n, bins, patches = plt.hist(value2, bins = 30, color='green') # hitt함수가 3개의 값읇 반환함 (빈, 빈들의 edge, 패치)
print("1. bins : " + str(bins)) # bins 변수는 bin의 edge. 즉, 모서리를 의미한다. 여기서 bin 개수가 30개이기 때문에 31개의 edge가 존재하게 된다.(!!). 두 개의 자료형을 하나의 자료형으로 통일시킴
print("2. the length of bins : " + str(len(bins))) # bins의 데이터 개수를 구함


    # 각 빈의 중간값을 구하기

bin_centers = 0.5 * (bins[:-1] + bins[1:]) # 배열형태의 bins 변수 안에 있는 값들을 인덱싱
                                                            # bins[:-1] : bins 변수 첫번째 ~ 마지막 요소 전까지 담고 있는 배열
                                                            # bins[1:] : bins 변수 두번째 ~ 마지막 요소까지 담고 있는 배열
print("3. bin_centers :" + str(bin_centers))

    # scale values to interval[0, 1] : 정규화(normalization)
col = (bin_centers - min(bin_centers)) / (max(bin_centers) - min(bin_centers)) # 데이터 전체 구간을 정해진 범위 안으로 변환하여 데이터를 관찰하는 방법
                                                                                                                 # 최소-최대 정규화 : 새로운값 = (현재값  - chlthrkqt ) / (최대값 - 최소값)
print("4. col : " + str(col))

for c, p in zip(col, patches): # 변수 개수만큼 pylot의 setp() 함수 호출
    plt.setp(p, 'facecolor', cm(c)) # cm(c) : RGBA 값 형태로 추출

plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram 3')
plt.show()


    


```

```python
###### Seaborn (Statistical Data Visulalization)
import seaborn as sns
    # Matplotlib를 기반으로 하는 파이썬 패키지. 통계적 그래픽을 제공하는 고차원 인터페이스 제공
    
    
# 히트맵 그리기 (색상으로 표현할 수 있는 정보를 열 분포 형태로 나타낸 그래프)

    # 1. 내장 데이터셋 "flights" 로드
data2 = sns.load_dataset("flights")
data2

    # 2. 데이터 pivot 시키기
        # index : "year"
        # column : "month"
        # value : "passengers"
pivoted_data = data2.pivot("year", "month", "passengers")
pivoted_data
        

    # 3. 스타일 옵션 변경하여 히트맵으로 나타내기
        # set() : 시각화 옵션 변경
sns.set(context="poster", font="monospace")
        # pivoted_data : 사용할 데이터
        # annot : 히트맵의 각 셀에 데이터의 값을 표시할 것인지 설정
        # fmt : "d"로 설정하면 10진수 형태로 출력
        # linewidth : 선두께
sns.heatmap(pivoted_data, annot=True, fmt='d', linewidth=1)

    # 4. 옵션 초기화 하기 seaborn 패키지 사용 후 matpotlib을 사용하여 시각화를 하게 되면 seaborn에 사용했던 기본 세팅이 그대로 유지가 되는 현상이 발생
sns.set()
sns.reset_orig()
sns.heatmap(pivoted_data, annot=True, fmt="d", linewidth=3)
    
# sns.set(style='whitegrid', palette = 'pastel')

# data = sns.load_dataset("tips")
# data


# sns.boxplot(x="day", y="tip", hue="sex", data=data, palette="muted")
























```
