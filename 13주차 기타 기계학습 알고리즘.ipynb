# K-means 알고리즘
비지도학습의 한 종류인 클러스티랑의 대표적인 알고리즘
 - 비지도 학습(unsupervised learning) : 데이터들과 각 클러스트와의 거리 차이의 분산을 최소화한다.
 - 클러스터링(군집) : 라벨 데이터 없이 데이터 각각의 특성을 고려해 주어진 데이터를 가장 잘 설명하는 집단을 찾아 속하게 하는 것

K-means 알고리즘은 주어진 데이터를 K개의 클러스터로 묶는 알고리즘이다. 이때 클러스터를 나누는 기준은 유사도를 기반으로 데이터들 간의 유사성을 판단한다.

## 원리
1. 클러스터 내 응집도를 최소화
  ![스크린샷, 2017-11-23 17-23-32](https://i.imgur.com/iURwnLE.png)
 - G1 데이터와 C1과의 거리 합 최소화
 - G2 데이터와 C2와의 거리 합 최소화
 - 여기서 중요한 개념이 나오는데, 앞에서 데이터 간의 거리 측정을 통해서 유사도를 측정한다고 설명했다. 여기서 거리 측정은 `센트로이드`라고 부르는 클러스터의 중심점과 각 데이터 간에 이루어진다.
 - 클러스터의 중심점은 말 그대로 클러스터 내의 데이터와 클러스터 중심점 간의 거리가 최소가 되는 점이고 `여러 번의 학습`을 통해서 이 중심점을 찾게 된다.
2. 클러스터 간 분리도 최대화
![스크린샷, 2017-11-23 17-24-32](https://i.imgur.com/dsODsKT.png)
 - G1의 중심 값(C1)과 G2의 중심 값(C2)과의 거리 최대화
 - 두 번째로 클러스터 간 분리도를 최대화 한다. 이는 다른 클러스터 간의 중심점의 거리가 최대화되어야 한다는 것이다. 그래야 각각의 클래스 구분이 쉬워진다.

## 거리측정법

### 유클리드 거리 측정법
![스크린샷, 2017-11-23 17-48-10](https://i.imgur.com/sPPdT3w.png)
유클리드 거리가 큰 값을 가지면 사용자 간의 거리가 멀다는 의미이기 때문에 사용자 간 유사성이 떨어짐. 벡터 간의 차이를 제곱한 합계 루트 값으로 거리를 측정한다.

### K-means 알고리즘 수행 단계
1. 초기 중심값 선택
   - 사용자가 k값을 임의로 정한다.
   - k개의 점을 임의로 선택해 k개의 클러스터의 초기 중심값을 정한다.

2. 클러스터 할당
   - k개의 클러스터의 중심 값과 각 개별 점(데이터)간의 거리를 측정하여 가장 가까운 클러스에 해당 데이터를 할당

3. 새로운 중심 값 선택
   - 각 클러스터마다 그 안에 배정된 모든 점들 간의 거리 평균값을 구하여 새로운 중심 값으로 정함.

k개의 중심점이 변하지 않을 떄까지 위 세 단계를 반복하게 되고, 어느 정도의 오차 범위 안에 들어오게 되면 반복을 중단한다. (즉, 새로운 중심 값들이 이전의 중심값들과 일치하지 않으면 반복)

# 예제1

````python
# 1. Python  패키지 가져오기 및 matplotlib 출력 옵션 설정

from sklearn.cluster import KMeans #  K-Means 클러스터링을 위한 함수
from sklearn.datasets import make_blobs # 가우시안 값들로 이루어진 샘플 데이터 셋을 생성하느 ㄴ함수

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
````

````python
# 2. 클러스터링 샘플 데이터 생성

np.random.seed(0)
    # numpy.random.seed() : 임의의 난수 값들을 생성하기 위한 시드(seed) 값 설정 함수. seed를 기반으로 난수가 만들어진다.
    # 시드(seed) : 임의의 난수 생성 시 사용되는 초기값 
        # 같은 시드 값을 설정할 경우 이전과 동일한 난수 값 생성
        # 시드 값이 설정되어 있지 않으면 생성할 떄마다 다른 값들이 생성된다.

centers = [[1, 1], [0, 0], [2, -1]] # 3개의 중심축을 기준으로 하는 데이터 생성, 

data, labels_true = make_blobs(n_samples=2000, centers=centers, cluster_std = 0.7)
    # make_blob() : 클러스터링 모델을 시험할 때 원하는 특성을 가진 데이터를 만들기 위해 사용하는 함수
    # centers=centers : make_blobs() 함수로 생성할 데이터 셋의 중심점들
    # cluster_std = 0.7 : 각 군집의 표준편차 값을 0.7로 설정
        # 표준편차 : 데이터의 산포도를 나타내는 수치로, 클수록 중심점으로부터 데이터들의 거리가 멀어지게 된다.

data.shape
````

````python
# 3. 데이터 및 라벨 확인
data.shape

print(data)
print("============================")
print(labels_true) # 2,000개의 좌표 값이 처름에 소속되는 클러스터 번호
print(np.unique(labels_true)) # 네 번째 줄에서 numpy의  unique() 함수를 사용해서 labels_true 변수에서 중복을 제거한 값을 출력 (즉, 생성한 데이터들이 각각 0, 1, 2 클러스터에 속한다.)
````

````python
# 4. 차원 평면에 데이터 나타내기
plt.figure(figsize=(15,10)) 
    # figure 함수로 가로 15, 세로 10으로 그림 크기를 설정한다.

plt.scatter(data[ : , 0], data[ : , 1])
    # 산포도로 나타낼 데이터의 x축 값, y축 값을 각각 파라미터로 넣어서 호출한다.
    # X축값 : 첫 번째 feature 값
    # Y축값 : 두 번째 feature 값
````

````python
# 5. 모델 학습시키기

estimator = KMeans(init='k-means++', n_clusters=3, n_init=10)
    # init = 'k-means++' :  KMeans 클러스터링 알고리즘의 초기 k값을 선택하는 알고리즘
    # n_clusters = 3 : 클러스터링 할 개수 3개
    # n_init = 10 :  초기 중심점을 선택할 때 최상의 결과를 얻기 위해서 몇 번 초기값을 변경할 것인지를 설정하는 파라미터
estimator.fit(data)
     # kmeans 초기 모델을 estimator 변수에 할당한 뒤에 fit() 함수를 사용해 data를 클러스터링 한다.
````

````python
# 6. 클러스터링 결과 확인하기

labels_predict = estimator.labels_
    # fit 함수를 호출하게 되면, estimator 변수는 labels_ 라는 클래스 변수에 클러스터 결과를 저장한다.
    # 이 값을 labels_predict 변수에 저장하고, numpy의 unique 함수를 사용해서 중복을 제거한다.
print(type(labels_predict))
print("labels_predict : ", labels_predict)

np.unique(labels_predict)
````

````python
# 6. 클러스터링 결과 확인하기

cm = plt.cm.get_cmap('jet')
    # matplotlib에서 제공하는 colormap 사용
    # 0~1 사이의 값을 인자로 받아 그에 상응하는 RGBA 값을 반환한다.
    # colormap 코드 중 하나인 'jet'를 pyplot.cm.get_cmap() 함수로 가져와서 cm 변수에 저장한다.
    
scaled_labels = (labels_predict - np.min(labels_predict))
    # 정규화 작업진행
    # labels_predict 변수에서 최소값을 뺀 값을 저장

scaled_labels = scaled_labels / (np.max((labels_predict) - np.min(labels_predict)))
    # 정규화 작업진행
    # 위에서 저장한 scaled_labels 변수를 최대값에서 최소값을 뺀 값으로 나누어서 다시 qustndp wjwkd.

np.unique(scaled_labels) # 정규화되어서 라벨값이 0, 1, 2에서 0, 0.5, 1로 정규화된다.
````

````python
plt.figure(figsize=(15,10)) # 그림의 크기 지정
plt.scatter(data[:, 0], data[:, 1], c=cm(scaled_labels)) 
    # scatter 함수 호출.
    # 이번에는 산포도에 색상을 입히기 위해 'c' 파라미터를 사용한다. 
````

# 예제2
init 함수를 다르게 해서 결과를 비교하고, 몇 가지 지료르 통해 클러스터링 성능 평가를 진행하는 예제2이다.

````python
# 파이썬 패키지 가져오기
from sklearn.datasets import load_digits # 0~9까지의 정수로 이루어진 샘플 데이터 셋을 가져오는 함수
from sklearn.preprocessing import scale 
    # 데이터 전처리를 위한 scale 함수
    # scale을 통해서 값의 규모를 조정해본다.
    # 데이터는 여러 feature들은 각각 값의 범위가 제 각각이다.
    # EX) 나이, 몸무게, 월급 등은 값의 범위가 크게 서로 다르다. => 모델의 성능에 영향을 줄 수 있다.
    # 왜냐하면, 데이터 간의 거리를 기반으로 만들기 때문에, 특정 feature값의 범위가 크면 다른 feature들보다 모델에 더 많은 영향을 주기 때문이다.
    # 그래서 모든 feature가 모델 생성 및 학습에 되대한 동일학 게이려르 하기 위해서 모델 생성 전 스케일링 같은 데이터 전처리 과정을 필수적으로 거치는 것이 좋다.
    # scale() : 각 데이터의 feature 값의 분포를 평균 0, 분산 1로 만든다.
import pandas as pd 

np.random.seed(60)
````

````python
# 1. 샘플 데이터 생성하기 및 데이터, 라벨 확인하기

digits = load_digits() 
data = digits.data

print("< Before scaling >")
print(data)

print("=====================================")

print("< After scaling >")
data = scale(data) # 스케일링을 하면 데이터 셋의 평균이 0, 분산이 1이 되도록 만들어진다.
print(data)
````

````python
# 2. 데이터 개수 및 feature, label 개수 확인하기
labels_true = digits.target # data 변수의 target 값을 저장한다.
n_samples, n_features = data.shape # data 변수의 shape 값을 가져와서 각각 n_samples와 n_feature에 저장. (data 변수의 행의 개수가 n_sampels 변수에, 열의 개수가 n_feautre에 저장)
clusters = np.unique(labels_true) # 실제 라벨 값들을 clusters라고 하는 변수에 저장한다.
n_clusters = len(clusters)  # 클러스터의 개수를 len() 함수를 이용해 n_clusters 변수에 할당한다.

print("n_samples : " + str(n_samples)) # 총 1797개의 데이터
print("n_features : " + str(n_features)) # 64개의 feature
print("n_clusters : " + str(n_clusters)) # 0부터 9까지 10개의 클러스터로 나뉜다.
print("clusters : " + str(clusters)) 
````

````python
# 3.초기 k 값 설정을 위한 nit 함수를 달리하여 각각 모델 학습시키기
estimator1 = KMeans(init = 'k-means++', n_clusters = n_clusters, n_init = 10) # k-means++ 함수 사용
estimator1.fit(data)

estimator2 = KMeans(init = 'random', n_clusters = n_clusters, n_init = 10) # random 함수 사용
estimator2.fit(data)

labels_predict1 = estimator1.labels_ # 두 모델을 사용해서 클러스링 한 데이터의 클러스터 라벨 값을 각각 labels_predict1과 labels_predict2 저장.
labels_predict2 = estimator2.labels_
````

# Scikit-Learn을 이용한 KNN 연습

````python
import sklearn.datasets as datasets
from sklearn.neighbors import KNeighborsClassifier # KNN 패키지
from sklearn.metrics import accuracy_score # 정확도 측정 관련 패키지
from sklearn.cross_validation import train_test_split # 교차 검증을 위한 패키지
from sklearn.model_selection import cross_val_score # 교차검증 정확도를 위한 패키지

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

````python
from sklearn.datasets import load_iris # 내장데이터로부터 iris data를 불러들임
iris = load_iris()
X = iris.data # train set으로 iris.data의 튜플값을 할당 (독립변수)
y = iris.target # target 값으로 iris.target의 튜플값을 할당 (종속변수)
````

````python
df = pd.DataFrame(X, columns = iris.feature_names)
print("< Iris Data >")
print("The number of sample data : " + str(len(df)))
print("The number of features of the data : " + str(len(df.columns)))
print("The labels of the data : " + str(np.unique(y))) # target값의 unique. (setosa, versicolor, virginia)
df
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
    # X_train : 현재 네 개의 값에 온 칼럼 값을 할당
    # y_train : label을 할당
    # random_state : 숫자가 주어지지 않으면 훈련 및 테스트 데이터가 매번 다르게 선택된다.
    #                         숫자가 주어지면 해당 숫자를 기반으로 랜덤 넘버 생성
    # 여기서는, 숫자를 주지 않으면 동일한 조건에서 성능평가가 어려워지기 때문에 숫자를 부여했다.
    
print("The number of train data set : %d " %len(X_train)) # 100개 (66%)
print("The number of test data set : %d " %len(X_test))   # 50개 (33%)
````

````python
estimator = KNeighborsClassifier(n_neighbors=3)
    # instantiate learning model (k = 3)
    # 내 주변에 몇 개의 눈치를 보고, 다수결로 나의 클래스를 정할 것인가? 즉, 3개를 보고 결정하라는 이야기.

estimator.fit(X_train, y_train)
    # fitting the model (모델을 만든다)

label_predict = estimator.predict(X_test) 
    # 독립변수가 X_test가 되고, 결과값(종속변수)가 label_predict로 할당된다.
    # predict the response (모델을 만든 것 가지고 예측)


print("The accuracy score of classification: %.9f"  
      %accuracy_score(y_test, label_predict)) 
        # 실제 나와야 할 값(y_test), 예측결과(label_predict)
        # evaluate accuracy (정확도 평가)

# 결과는 98%가 나오는데, 과최적화의 문제가 있다. 그래서 많이 쓰이는 방법이 10-fold cross validation이다.
````

````python
# 전체 데이터를 10등분하고, 1/10을 test set으로 쓰고, 나머지 9/10을 train set으로 쓴 다음에 모델을 만들고 결과값을 본다.

myList = list(range(1,50)) # create odd list of k for KNN (홀수를 찾아냄)
neighbors = [ x for x in myList if x % 2 != 0]
print(neighbors)
print("The number of neighbors k is %d" %len(neighbors))
````

````python
# empty list that will hold cross validation scores
cv_scores = [] # 결과값의 정확도를 보고자 함. 결과값의 정확도를 담는다.

# perform 10-fold cross validation
for k in neighbors:
    print("< k = %d >" %k)
    estimator = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(estimator, X_train, y_train, cv = 10, scoring = 'accuracy')
        # X_train : training data(독립변수)
        # y_train : training data의 종속변수
        # cv = 10 : 10등분 할 것 (9개는 training set, 1개는 test set)
    
    print("The scores of classification are \n" + str(scores))
    cv_scores.append(scores.mean()) # average error  (10등분 한 것의 첫 번째 fold에 대해서 테스트,  두 번째 fold에 대해서 테스트, ... )
    print("The average score of scores is %.9f \n" %scores.mean())
    
    # 1이 나오면 과최적화가 되었다는 뜻. 그래서 데이터를 나누어서 측정하는 것이 중요하다
    # k=1, 3, 5, .... , 49까지 반복해서 그림을 그려본다.
````

````python


# changing to misclassification rate (a.k.a classification error)
# MSE = 1 - cross validation score
MSE = [1 - x for x in cv_scores]

# plot misclassification error vs k
plt.plot(neighbors, MSE)
plt.xlabel("Number of Neighbors K")
plt.ylabel("Misclassification Error")
plt.show()

# determining best k 
min_MSE = min(MSE)
index_of_min_MSE = MSE.index(min_MSE)
optimal_k = neighbors[index_of_min_MSE]
print ("The optimal number of neighbors i is %d" % optimal_k)

# error의 값이 높ㅇ르 수록 나쁘고, 적을 수록 좋다.
# k = 7 일때는 0.048일때 에러가 제일 작다.
````

````python
# instantiate learning model (k = 7일떄 최적의 값을 갖는다)
estimator = KNeighborsClassifier(n_neighbors=3)

# fitting the model
estimator.fit(X_train, y_train)

# predict the response
label_predict = estimator.predict(X_test)

# evaluate accuracy
print("The accuracy score of classification: %.9f"  
      %accuracy_score(y_test, label_predict))

# 결과값이 0.98이 나오는데, 앞에 결과값과 거의 비슷하게 나왔다.
````

````python
# perform 10-fold cross validation 

# create odd list of k for kNN
myList = list(range(1,100))
neighbors = [ x for x in myList if x % 2 != 0] # 홀수를 만들어서 실험
print(neighbors)
print("The number of neighbors k is %d" % len(neighbors)) # 전체 k는 50개가 생성된다.
````
